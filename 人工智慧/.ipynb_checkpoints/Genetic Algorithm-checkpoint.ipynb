```python
import numpy as np
import time
```

```python
# 單變數函數且線性
def F1(t):
    return 0.063*(t**3)-5.284*(t**2)+4.887*t+412+np.random.normal(0,1)
# 非線性產生器
def F2(t,A,B,C,D):  
    return A*(t**B)+C*np.cos(D*t)+np.random.normal(0,1,t.shape)# t.shape 一次產生一整個 columns 的亂數

# 產生空矩陣 用OLS 求解找出係數
A = np.zeros((1000,5))
b = np.zeros((1000,1))
for t in range(0,1000):
    b[t] = F1(t)
    A[t,0] = t**4
    A[t,1] = t**3
    A[t,2] = t**2
    A[t,3] = t
    A[t,4] = 1
x = np.linalg.lstsq(A,b)
print(x)
```

```python
# 已知確切的 form 如 F2 
# T 是產生均勻U(01)，1000個值*100變成均勻U(0,100)
T = np.random.random((1000,1))*100
# b2 是答案，理論上應該不知道，用下面的基因演速法在ABCD當中的range 找值
# 接著用 Fitness function 比較兩者之差找出最小的 Fitness function 就可以找到最佳 ABCD
b2 = F2(T,0.6,1.2,100,0.4)
```

```python
def gene2coef(gene): # 二進位轉十進位 共有2^40碼
    # 共1024個值，用10個01代替，1-1 mapping
    # 2**np.arange(10) 是 array([  1,   2,   4,   8,  16,  32,  64, 128, 256, 512])
    # *gene[0:10] 產生 10 個 01 
    # 相乘就會將二進位轉成十進位
    A = (np.sum(2**np.arange(10)*gene[0:10])-511)/100 #-5.11~5.12 
    B = (np.sum(2**np.arange(10)*gene[10:20])-511)/100 
    C = (np.sum(2**np.arange(10)*gene[20:30])-511)
    D = (np.sum(2**np.arange(10)*gene[30:40])-511)/100
    return A,B,C,D
```

```python
# 一代10000(N)人 作 10 (G)代
N = 10000
G = 30
# 存活與突變率
survive_rate = 0.05
mutation_rate = 0.001
survive = round(N*survive_rate)
mutation = round(N*40*0.001) #突變
# 亂數產生一萬組 01 的基因 N = 10000
pop = np.random.randint(0,2,(N,40))
fit = np.zeros((N,1))
```

```python
#適者生存-交配-突變
# 對每十代
%time
for generation in range(G):
    print(generation)
    # 對每一萬人
    ## 適者生存
    for i in range(N):   
        # 二進位轉十進位
        A,B,C,D = gene2coef(pop[i,:])
        # 比較兩者之差找出最小的 Fitness function
        # 越小越好
        fit[i] = np.mean(abs(F2(T,A,B,C,D)-b2))
    # selection
    # 適者生存從小到大
    sortf = np.argsort(fit[:,0]) #照大小排，15364->02413
    # 取前 N 個人
    pop = pop[sortf,:]
    # 從 N+1 到 最後 被覆蓋掉
    ##交配
    for i in range(survive,N):
        # 從 0-499取爸媽
        fid = np.random.randint(0,survive)
        mid = np.random.randint(0,survive)
        # 同值重取
        while(fid==mid):
            mid = np.random.randint(0,survive)
        # 產稱 40個01 0從爸爸 1從媽媽
        mask = np.random.randint(0,2,(1,40))
        # 媽媽基因複製給兒子
        son = pop[mid,:].copy()
        father = pop[fid,:]
        # 如果是==1 換成爸爸
        son[mask[0,:]==1] = father[mask[0,:]==1]
        pop[i,:] = son
    ## 突變 從9500個人有人會突變
    for i in range(mutation):
        m = np.random.randint(survive,N)
        n = np.random.randint(0,40)
        # 0變1 1變0
        pop[m,n] = 1-pop[m,n]
        
#整個演算法跑完看到第三十代
for i in range(N):
    A,B,C,D = gene2coef(pop[i,:])
    fit[i] = np.mean(abs(F2(T,A,B,C,D)-b2))
sortf = np.argsort(fit[:,0])
pop = pop[sortf,:]

#回傳第0個人
A,B,C,D = gene2coef(pop[0,:])
print(A,B,C,D)
```
