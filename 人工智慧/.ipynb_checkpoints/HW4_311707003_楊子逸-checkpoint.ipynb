# 已知 A=0.6、B=1.2、C=100，試以 D 值可能出現的 1024 種可能性為 X 軸，Energy(A,B,C,D)為 Y 軸，繪製折線圖

```python

def F2(t,A,B,C,D):  
    return A*(t**B)+C*np.cos(D*t)+np.random.normal(0,1,t.shape)# t.shape 一次產生一整個 columns 的亂數

# 已知確切的 form 如 F2 
# T 是產生均勻U(0,1)，1000個值*100變成均勻U(0,100)
T = np.random.random((1000,1))*100
# b2 是答案，理論上應該不知道，用下面的基因演速法在ABCD當中的range 找值
# 接著用 Fitness function 比較兩者之差找出最小的 Fitness function 就可以找到最佳 ABCD
b2 = F2(T,0.6,1.2,100,0.4)

def gene2coef(gene):    # 2 進制轉成 10 進制
    tc = (np.sum(2**np.arange(4)*gene[0:4])+500)
    beta = (np.sum(2**np.arange(10)*gene[4:14])+0.000000000000000001)/1024
    omega = (np.sum(2**np.arange(10)*gene[14:24])+0.000000000000000001)/1024*10
    # omega = math.log2(np.sum(2**np.arange(10)*gene[14:24])+2)+1
    # phi = (np.sum(2**np.arange(10)*gene[24:34]))/1024*2*math.pi
    phi = (np.sum(2**np.arange(10)*gene[24:34])+0.000000000000000001)/1024*2*math.pi
    return tc, beta, omega, phi
```

```python
b2 = F2(T,0.6,1.2,100,0.4)
T = np.random.random((1000,1))*100
energy = []
for i in np.arange(-5.12,5.12,0.01):
    e = np.mean(abs(F2(T,0.6,1.2,100,i))-b2)
    energy.append(e)
```

```python
df = pd.DataFrame()
df['range'] = np.arange(-5.12,5.12,0.01)
df['energy'] = energy
x = df['range'] # x軸的值
y = df['energy']  # y軸的值
plt.figure(figsize=(20, 10))
plt.title(" D value") # 圖的標題
plt.xlabel("x axis") # x軸的名稱
plt.ylabel("Energy(A,B,C,D)") # y軸的名稱
plt.plot(x,y) # 繪製x,y軸的圖
plt.show() # 顯現圖形
```

# 已知 tc 參數出現在 2021/11/24 與 2021/11/30 之間(含)，請用基因演算法來推測非線性 參數 tc, β, ω, Φ，每次計算此基因的 fitness 前，先以線性迴歸推論此基因對應的最佳 A,B,C 參數，並以七個參數計算 p(t0)~p(tc) 序列(合成序列)，最後 fitness 再計算此序列與真實 NVIDIA 股價(真實序列)的差異。請在作業報告中將你求出的最佳參數組合列出，並將合 成序列與直實序列用兩種顏色繪製於同一張折線圖中。

```python
x = pd.read_csv('/Users/imath/Desktop/交通大學/碩一上/人工智慧理論與實作/1102基因演算法與LPPL/NVDA.csv')
x = np.array(x['Close'])
np.save('NVDA', x)
x = x.reshape((len(x), 1))
plt.plot(x)
```

```python
def logPt(tc, A, B, C, beta, omega, phi):
    t = np.arange(0, tc).reshape((tc, 1))
    return A + B*(tc-t)**beta + B*C*(tc-t)**beta*np.cos(omega*np.log(tc-t)+phi)

def logPtFit(tc, A, B, C, beta, omega, phi, true):
    fitness = np.mean(abs(logPt(tc, A, B, C, beta, omega, phi) - true))
    return fitness

def MAE(real, predict):
    return np.mean(abs(real[:len(predict)]  - predict))   

def LR(tc, beta, omega, phi, b):
    AMatrix = np.zeros((tc, 3))
    t = np.arange(tc).reshape((tc, 1))
    AMatrix[t, 0] = 1
    AMatrix[t, 1] = (tc-t)**beta
    AMatrix[t, 2] = (tc-t)**beta*np.cos(omega*np.log(tc-t)+phi)
    return np.linalg.lstsq(AMatrix, np.log(b), rcond=None)[0]
```

```python
def gene2coef(gene):    # 2 進制轉成 10 進制
    tc = (np.sum(2**np.arange(4)*gene[0:4])+860)
    beta = (np.sum(2**np.arange(10)*gene[4:14])+0.000000000000000001)/1024
    omega = (np.sum(2**np.arange(10)*gene[14:24])+0.000000000000000001)/1024*10
    # omega = math.log2(np.sum(2**np.arange(10)*gene[14:24])+2)+1
    # phi = (np.sum(2**np.arange(10)*gene[24:34]))/1024*2*math.pi
    phi = (np.sum(2**np.arange(10)*gene[24:34])+0.000000000000000001)/1024*2*math.pi
    return tc, beta, omega, phi
```

```python

start = time.time()
popNum = 1000
popDim = 34
generationNum = 30
bestPopRate = 0.3
bestPopNum = int(popNum * bestPopRate)
mutationRate = 0.1
mutationNum = int(popNum * mutationRate)

pop = np.random.randint(0, 2, (popNum, popDim))
fit = np.zeros((popNum, 1))

b = x.copy().reshape((len(x), 1))

true_value = np.load('NVDA.npy')
result = 0.033262727098503524
flagTemp = 0
flagResult = 0


## 世代
for generation in range(generationNum):
    print('generation: ', generation)
    for i in range(popNum):
        tc, beta, omega, phi = gene2coef(pop[i, :])
        bMatrix = b[:tc]
        A, B, X = LR(tc, beta, omega, phi, bMatrix)
        C = X / B
        true = x[:tc]
        TEMP = logPt(tc, A, B, C, beta, omega, phi) - np.log(true)
        # print(logPt(tc, A, B, C, beta, omega, phi) - np.log(true))
        fit[i] = np.mean(abs(logPt(tc, A, B, C, beta, omega, phi) - np.log(true)))
    sortf = np.argsort(fit[:, 0])
    # print('fit: ', fit)
    pop = pop[sortf, :]
    
    ## 繁衍
    for i in range(bestPopNum, popNum):
        fid = np.random.randint(0, bestPopNum)
        mid = np.random.randint(0, bestPopNum)
        while mid == fid:    # 爸爸媽媽基因一樣就重取
            mid = np.random.randint(0, bestPopNum)
        mask = np.random.randint(0, 2, (1, popDim))    # 產生一組 0、1 遮罩
        son = pop[mid, :]    # 0 copy 媽媽、1 copy 爸爸
        father = pop[fid, :]
        son[mask[0, :] == 1] = father[mask[0, :] == 1]
        pop[i, :] = son
    
    ## 突變
    for i in range(mutationNum):
        m = np.random.randint(0, popNum)
        n = np.random.randint(0, popDim)
        pop[m, n] = 1 - pop[m, n]    # 0 變 1、1 變 0
    
    besttc, bestbeta, bestomega, bestphi = gene2coef(pop[0, :])
    bMatrix = b[:tc]
    A, B, X = LR(tc, beta, omega, phi, bMatrix)
    C = X / B
    true = x[:tc]
    roundbestfit = np.mean(abs(logPt(tc, A, B, C, beta, omega, phi) - np.log(true)))
    if roundbestfit < result:
        resulttc = tc
        resultA = A
        resultB = B
        resultC = C
        resultbeta = beta
        resultomega = omega
        resultphi = phi
        flagTemp = 1
        print('FLAGTEMP')

    print('round best: ', besttc, A, B, C, bestbeta, bestomega, bestphi)
    print('round best fit: ', roundbestfit)

## 最後一次更新數值
for i in range(popNum):
    tc, beta, omega, phi = gene2coef(pop[i, :])
    bMatrix = b[:tc]
    A, B, X = LR(tc, beta, omega, phi, bMatrix)
    C = X / B
    true = x[:tc]
    fit[i] = np.mean(abs(logPt(tc, A, B, C, beta, omega, phi) - np.log(true)))
sortf = np.argsort(fit[:, 0])
pop = pop[sortf, :]

## 取最好的一次結果
tc, beta, omega, phi = gene2coef(pop[0, :])
bMatrix = b[:tc]
A, B, X = LR(tc, beta, omega, phi, bMatrix)
C = X / B
true = x[:tc]
bestFit = np.mean(abs(logPt(tc, A, B, C, beta, omega, phi) - np.log(true)))
print(tc, A, B, C, beta, omega, phi)
print('bestFit: ', bestFit)


predict = logPt(tc, A, B, C, beta, omega, phi)
predictExp = np.exp(predict[:])

plt.plot(x)
plt.plot(predictExp)
plt.show()

end = time.time()
print('time: ', end - start)

if bestFit < result:
    resulttc = tc
    resultA = A
    resultB = B
    resultC = C
    resultbeta = beta
    resultomega = omega
    resultphi = phi
    flagResult = 1
    print('FLAGRESULT')

flagResult = 1
print('FLAGRESULT')
print(MAE(x,predictExp))
```
